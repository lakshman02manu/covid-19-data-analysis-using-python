```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
print('Modules are imported.')
```

```python
corona_dataset = pd.read_csv("datasets/covid19_Confirmed_dataset.csv")
corona_dataset.head(10)
```

```python
corona_dataset.shape
```

```python
df = corona_dataset.drop(labels=["Lat","Long"],axis=1)
```

```python
df.head(10)
```

```python
aggregate =  df.groupby("Country/Region").sum()
```

```python
aggregate.head()
```

```python
aggregate.shape
```

```python
aggregate.loc["India"].plot()
aggregate.loc["Pakistan"].plot()
aggregate.loc["China"].plot()
plt.legend()
```

```python
countries = list(aggregate.index)
highlyaffected = []
for c in countries :
    highlyaffected.append(aggregate.loc[c].diff().max())
aggregate["highlyaffected"] = highlyaffected
aggregate.head()
```

```python
infected = pd.DataFrame(aggregate["highlyaffected"])
infected.head()
```
